# Логистическая регрессия
<b>Логистическая регрессия</b> используется для моделирования бинарного исхода (например, да/нет, 0/1). Она предсказывает вероятность принадлежности примера к одному из двух классов.
### Пример:
Представим, что мы хотим предсказать, сдал ли ученик экзамен (да/нет) на основе количества часов, которые он учил предмет.
### Уравнение логистической регрессии:
 $$\hat{p} = \frac{1}{1 + e^{-(b_0 + b_1 \cdot x)}} $$
 	•	$\hat{p}$ — вероятность того, что ученик сдал экзамен.<br>
	•	$e$ — основание натурального логарифма.<br>
	•	$b_0$ и $b_1$ — параметры модели, аналогичные линейной регрессии.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
# Данные
hours = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
passed = np.array([0, 0, 1, 1, 1])
```

```python
# Создаем и обучаем модель
model = LogisticRegression()
model.fit(hours, passed)
```

```python
#Предсказываем вероятность сдачи экзамена
predict_probabilities = model.predict_proba(hours)[:, 1]
predict_probabilities
#model.predict_proba(hours)[:, 1] — возвращает вероятности принадлежности каждому из двух классов (0 и 1). 
#Мы используем [:, 1], чтобы получить вероятности для класса “сдал экзамен” (то есть вероятность того, что результат равен 1).
```

```python
model.predict_proba(hours)
```

```python
# Визуализирууем
plt.scatter(hours, passed, color='blue', label='Настоящие данные')
plt.plot(hours, predict_probabilities, color='red', label='Предсказанная вероятность')
plt.xlabel('Часы учебы')
plt.ylabel('Вероятность сдачи экзамена')
plt.title('Логистическая регрессия')
plt.legend()
plt.show()
```
