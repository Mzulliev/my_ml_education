# Линейная регрессия 
<b>Линейная регрессия</b>- это метод для моделирования отношений между зависимой переменной (целевая переменная) и одной или более независимыми переменными (признаки). Цель — найти линейное уравнение, которое наилучшим образом предсказывает целевую переменную.
### Уравнение линейной регрессии:
$$ \hat{y} = b_0 + b_1 \cdot x $$
	•	$\hat{y}$— предсказанное значение целевой переменной (оценка).<br>
	•	$x$ — значение независимой переменной (часы учебы).<br>
	•	$b_0$ — свободный член (пересечение с осью y).<br>
	•	$b_1$ — коэффициент наклона (показывает, как сильно влияет x на \hat{y}).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
hours = np.array(range(1, 6)).reshape(-1, 1)
scores = np.array([2, 4, 5, 4, 5])
```

```python
# создаем и обучаем модель
model = LinearRegression()
model.fit(hours, scores)
```

```python
print('Коэффициент наклона', model.coef_[0]) #коэффициент наклона
print('Свободный член', model.intercept_) #свободный член
```

```python
# предсказываем оценки
predicted_scores = model.predict(hours)
```

```python
#настоящие данные
plt.scatter(hours, scores, color='blue') 
# предсказания
plt.plot(hours, predicted_scores, color='red')
plt.xlabel('Чаcы учебы')
plt.ylabel('Оценка')
plt.title('Линейная регрессия')
plt.show()
```
